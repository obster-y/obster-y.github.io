# NETWORK

```python
# CLASS: ReplayBuffer
class ReplayBuffer:
    """Fixed-size buffer to store experience tuples."""

    def __init__(self, action_size, buffer_size, batch_size, seed):
        """Initialize a ReplayBuffer object.
        Params
        ======
            action_size (int): dimension of each action
            buffer_size (int): maximum size of buffer
            batch_size (int): size of each training batch
            seed (int): random seed
        """
        self.action_size = action_size
        self.memory = deque(maxlen=buffer_size)
        self.batch_size = batch_size
        self.experience = namedtuple("Experience", field_names=[
                                     "state", "action", "reward", "next_state"])
        self.seed = random.seed(seed)

    def add(self, state, action, reward, next_state):
        """Add a new experience to memory."""
        e = self.experience(state, action, reward, next_state)
        self.memory.append(e)

    def sample(self):
        """Randomly sample a batch of experiences from memory."""
        experiences = random.sample(self.memory, k=self.batch_size)

        states = torch.from_numpy(
            np.vstack([e.state for e in experiences if e is not None])).float().to(DEVICE)
        actions = torch.from_numpy(
            np.vstack([e.action for e in experiences if e is not None])).long().to(DEVICE)
        rewards = torch.from_numpy(
            np.vstack([e.reward for e in experiences if e is not None])).float().to(DEVICE)
        next_states = torch.from_numpy(np.vstack(
            [e.next_state for e in experiences if e is not None])).float().to(DEVICE)

        return (states, actions, rewards, next_states)

    def __len__(self):
        """Return the current size of internal memory."""
        return len(self.memory)
```

```python
class QNetwork(nn.Module):
    """DNN Model"""

    def __init__(self, state_size=1, action_size=7, seed=1, lstm_hiddensize=128,
                 fc1_delta=1, fc2_units=128, fc3_units=64):
        """Initialize parameters and build model.
        Params
        ======
            state_size (int): Dimension of each state
            action_size (int): Dimension of each action
            seed (int): Random seed
            fc1_units (int): Number of nodes in first hidden layer
            fc2_units (int): Number of nodes in second hidden layer
        """
        super(QNetwork, self).__init__()
        self.seed = torch.manual_seed(seed)
        self.lstm1 = nn.LSTM(input_size=state_size,
                             hidden_size=lstm_hiddensize)
        self.fc1 = nn.Linear(fc1_delta+lstm_hiddensize, fc2_units)
        self.fc2 = nn.Linear(fc2_units, fc3_units)
        self.fc3 = nn.Linear(fc3_units, action_size)

    def forward(self, state):
        """Build a network that maps state -> action values."""
        _, x = self.lstm1(state[:-1, :, :])
        x = Funcs.relu(
            self.fc1(torch.cat([x[0][0], state[0, :, -1].unsqueeze(dim=1)], dim=1)))
        x = Funcs.relu(self.fc2(x))
        return self.fc3(x)
```

```python
EV_EMAX = 24  # kWh
ACTION_VAL = np.linspace(6, -6, 7)  # Action > 0: charging, vice versa

ANXIETY_COEF = 0.01  # TAU
C_E = np.array([400,300,200,100,80,60,40,20])  # TODO
M_k = -0.015  # TODO


class RL_env(gym.Env):
    def __init__(self, data, n=32):
        self.action_space = spaces.Discrete(7)
        self.observation_space = None
        self.state = None
        self.sub_day = 0  # selected
        self.n = n

        self.data = data
        self.arrival_time_gen = self.truncnorm((18, 1, 15, 21))
        self.departure_time_gen = self.truncnorm((8, 1, 6, 11))
        self.soc_on_arrival_gen = self.truncnorm((0.5, 0.1, 0.2, 0.8))

        self.start_day = random.randint(0, (self.data.size - self.n * 24) / 24)
        self.arrive_series = np.rint(self.arrival_time_gen.rvs(n)).astype(int)
        self.depart_series = np.rint(
            self.departure_time_gen.rvs(n)).astype(int)
        self.soc_series = self.soc_on_arrival_gen.rvs(self.n) * EV_EMAX
        self.price_series = self.data[self.start_day *
                                      24:(self.start_day + self.n) * 24]

    def truncnorm(self, x):
        return stats.truncnorm((x[2] - x[0]) / x[1], (x[3] - x[0]) / x[1], loc=x[0], scale=x[1])

    def step(self, action, state, t, meta, battery_degrading):
        delta_e = ACTION_VAL[action]
        # - C_E * np.abs(M_k / 100) * delta_e / EV_EMAX
        state_next = state.copy()
        state_next[-1] += delta_e
        reward = -state[-2] * delta_e
        if t == 0:
            reward -= ANXIETY_COEF * (EV_EMAX - state_next[-1])**2
        else:
            pass

        if battery_degrading:
            reward -= C_E[action] * np.abs(M_k / 100) * delta_e / EV_EMAX

        state_next[:24] = meta[2][-t-24:-t]
        return state_next, reward

    def reset(self):
        if self.sub_day == self.n - 2:
            self.start_day = random.randint(
                0, (self.data.size - self.n * 24) / 24)
            self.arrive_series = np.rint(
                self.arrival_time_gen.rvs(self.n)).astype(int)
            self.depart_series = np.rint(
                self.departure_time_gen.rvs(self.n)).astype(int)
            self.soc_series = self.soc_on_arrival_gen.rvs(self.n) * EV_EMAX
            self.price_series = self.data[self.start_day *
                                          24:(self.start_day + self.n) * 24]
            self.sub_day = 0
        self.sub_day += 1
        _s = (self.arrive_series[self.sub_day],
              self.depart_series[self.sub_day], self.soc_series[self.sub_day])
        self.meta = (24 + _s[1] - _s[0], _s[2], self.price_series[(
            self.sub_day - 1) * 24 + _s[0]:(self.sub_day + 1) * 24 + _s[1]])
        # print(self.meta[0],self.meta[2].size,self.sub_day)
        return self.meta

    def close(self):
        return None
```

```python
BUFFER_SIZE = int(1e5)  # replay buffer size
BATCH_SIZE = 32  # F
LEARN_INTERVAL = 4
UPDATE_INTERVAL = 500  # B
INTERPOTLATION_COEF = 1
GAMMA = 0.99
LR = 5e-4               # learning rate
EPS = 0.01
time_str = time.strftime("%Y-%m-%d-%H-%M-%S", time.localtime(time.time()))


class AGENT():
    def __init__(self, state_size, action_size, seed, flg, time_str):
        """Initialize an Agent object.

        Params
        ======
            state_size (int): dimension of each state
            action_size (int): dimension of each action
            seed (int): random seed
        """
        self.state_size = state_size
        self.action_size = action_size
        random.seed(seed)
        # self.seed = random.seed(seed)

        # Q-Network
        self.qnetwork_local = QNetwork(
            state_size, action_size, seed).to(DEVICE)
        self.qnetwork_target = QNetwork(
            state_size, action_size, seed).to(DEVICE)
        self.optimizer = optim.Adam(self.qnetwork_local.parameters(), lr=LR)

        # Replay memory
        self.memory = ReplayBuffer(action_size, BUFFER_SIZE, BATCH_SIZE, seed)
        # Initialize time step (for updating every UPDATE_EVERY steps)
        self.step_t = 0
        self.step_b = 0

        self.time_str = time_str
        self.flg = flg

    def act(self, state, eps=EPS):
        """Returns actions for given state as per current policy.

        Params
        ======
            state (array_like): current state
            eps (float): epsilon, for epsilon-greedy action selection
        """
        state = torch.from_numpy(state).float()[:, None, None].to(DEVICE)
        self.qnetwork_local.eval()
        with torch.no_grad():
            action_values = self.qnetwork_local(state)
        self.qnetwork_local.train()

        # Epsilon-greedy action selection
        if random.random() > eps:
            return np.argmax(action_values.cpu().data.numpy())
        else:
            return random.choice(np.arange(self.action_size))

    def step(self, state, action, reward, state_next):
        # Save experience in replay memory
        self.memory.add(state, action, reward, state_next)

        # Learn every UPDATE_EVERY time steps.
        self.step_t = (self.step_t + 1) % LEARN_INTERVAL
        if self.step_t == 0:
            # If enough samples are available in memory, get random subset and learn
            if len(self.memory) > BATCH_SIZE:
                experiences = self.memory.sample()
                self.learn(experiences, GAMMA)

    def learn(self, experiences, gamma):
        """Update value parameters using given batch of experience tuples.
        Params
        ======
            experiences (Tuple[torch.Tensor]): tuple of (s, a, r, s', done) tuples 
            gamma (float): discount factor
        """
        states, actions, rewards, states_next = experiences
        states_next = states_next.T.unsqueeze(-1)
        # print(states_next.size())
        states = states.T.unsqueeze(-1)
        # Get max predicted Q values (for next states) from target model
        # Q_targets_next = self.qnetwork_target(states_next).detach().max(1)[1]
        if self.flg:
            Q_targets_next = self.qnetwork_target(
                states_next).detach().max(1)[0].unsqueeze(1)
        else:
            actions_next = self.qnetwork_local(
                states_next).detach().argmax(1).unsqueeze(1)
            Q_targets_next = self.qnetwork_target(
                states_next).detach().gather(1, actions_next)  # .unsqueeze(1)
        # Compute Q targets for current states
        Q_targets = rewards + (gamma * Q_targets_next)
        # Get expected Q values from local model
        Q_expected = self.qnetwork_local(states).gather(1, actions)

        # Compute loss
        loss = Funcs.mse_loss(Q_expected, Q_targets)
        # Minimize the loss
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        # ------------------- update target network ------------------- #
        self.step_b += 1
        self.step_b %= UPDATE_INTERVAL
        if self.step_b == 0:
            for p_target, p_local in zip(self.qnetwork_target.parameters(), self.qnetwork_local.parameters()):
                p_target.data.copy_(
                    INTERPOTLATION_COEF*p_local.data + (1.0-INTERPOTLATION_COEF)*p_target.data)

    def save(self, epoch):
        torch.save(self.qnetwork_local.state_dict(), PATH +
                   'LOCAL-{}-{}.tar'.format(epoch, self.time_str))
        torch.save(self.qnetwork_target.state_dict(), PATH +
                   'TARGET-{}-{}.tar'.format(epoch, self.time_str))

    def load(self, epoch):
        self.qnetwork_local.load_state_dict(
            torch.load(PATH+'EP_{}-LOCAL-CK.tar'.format(epoch)))
        self.qnetwork_target.load_state_dict(
            torch.load(PATH+'EP_{}-TARGET-CK.tar'.format(epoch)))
```

# TRAINING

```python
%matplotlib inline
def train(EPOCH=50000, eps_decay_start_epoch=7000, eps_decay_end_epoch=14000, eps_start=1.0, eps_end=0.1, target=True, bd = True):
    scores = []
    scores_last = deque(np.zeros(100),maxlen=100)
    scores_last_x = np.linspace(1,100,100)
    scores_last_win = vis.line(X=scores_last_x, Y=scores_last_x)
    scores_avg = []
    scores_avg_win = vis.line(X=np.arange(1),Y=np.arange(1))
    scores_avg_i = 0
    eps = eps_start
    eps_iter = np.nditer(np.concatenate([np.ones(eps_decay_start_epoch,), np.linspace(
        eps_start, eps_end, eps_decay_end_epoch-eps_decay_start_epoch), eps_end * np.ones(EPOCH-eps_decay_end_epoch)]))
    for ep in range(EPOCH):
        meta = env.reset()
        score = 0
        state = np.concatenate(
            [meta[2][-meta[0]-24:-meta[0]], np.array([meta[1], ])])
        for t in range(meta[0]-1, 0, -1):
            action = agent.act(state, eps)
            state_next, reward = env.step(action, state, t, meta, bd)
            agent.step(state, action, reward, state_next)
            state = state_next
            score += reward
        scores.append(score)
        scores_last.append(score)
        # eps = max(eps_end, eps_decay*eps) if ep >= eps_decay_start_epoch else eps_start  # decrease epsilon
        eps = next(eps_iter)
        if ep % 100 == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(
                ep, np.mean(scores_last)))
#             writer.add_scalar('scores_last', scores_last, global_step=None, walltime=None)
            vis.line(X=scores_last_x,Y=np.array(scores_last),win=scores_last_win,update='replace',opts=dict(title='{} / {}'.format(ep,EPOCH)))
            scores_avg.append(np.mean(scores_last)-2)
            scores_avg_i += 1
            vis.line(X=np.array([scores_avg_i]),Y=np.array([np.mean(scores_last)-2]),win=scores_avg_win,update='append')
        if ep % 5000 == 0:
            agent.save(epoch=ep)
            pass

    return (scores, agent)
```
